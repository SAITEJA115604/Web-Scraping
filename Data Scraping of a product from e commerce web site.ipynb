```python
from bs4 import BeautifulSoup as bs
```

```python
page = requests.get("https://www.flipkart.com/tyy/4io/~cs-23pjn3d2my/pr?sid=tyy,4io&collection-tab-name=iphone+13&param=44321")
```

```python
soup = bs(page.content,'html.parser')
```

```python
print(soup.prettify())
```

```python
name=soup.find('div',class_="col col-7-12")
print(name.prettify())
```

```python
name.text
```

```python
rating=soup.find('div',class_="gUuXy-")
print(rating.prettify())
```

```python
rating.text
```
